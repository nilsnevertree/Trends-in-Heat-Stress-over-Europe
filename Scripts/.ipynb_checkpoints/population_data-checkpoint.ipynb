```python
# calculate data_population with nearest option of surrounding places.

def calc_distance(lat1, lat2, lon1, lon2):
    '''
    NOTE: works with radians
    '''

    deg_to_rad = lambda x : x/180 *np.pi
    delta_lat = deg_to_rad(lat1) - deg_to_rad(lat2)
    delta_lon = deg_to_rad(lon1) - deg_to_rad(lon2)
    mean_lat = (deg_to_rad(lat1) + deg_to_rad(lat2))/2
    
    return (6371.009*1e3)**2*(delta_lat**2 + (np.cos(mean_lat)*delta_lon)**2)

# def poibilities(lat_idx,lon_idx):
    
    
    
def regridd_with_nearest(data, population):
    mask = mc.mask_land(data.t2m.isel(time = 0))
    lon_array, lat_array = np.meshgrid(data.longitude, data.latitude)
    lat_array[~mask.values] = np.nan
    lon_array[~mask.values] = np.nan
    lat_array = xr.DataArray(lat_array, dims= dict(latitude = data.latitude, longitude = data.longitude))
    lon_array = xr.DataArray(lon_array, dims= dict(latitude = data.latitude, longitude = data.longitude))
    
    plt.contourf(data.longitude, data.latitude,  lat_array)
    print(lat_array)
    plt.colorbar()
    plt.figure()
    plt.contourf(data.longitude, data.latitude,  lon_array)
    plt.colorbar()
        
    count = 0
    count1 = 0
    count2 = 0
    # calculate data_population:
    total_new = data.t2m.isel(time = 0) * 0
    print(population.total.sum().values)

    for lat_idx in range(len(total_new.latitude)) :
        lat_value = total_new.latitude[lat_idx].values
        lat_slice = slice(lat_value + 1.124/2, lat_value - 1.125/2)    
        
        for lon_idx in range(len(total_new.longitude)) :
            lon_value = total_new.longitude[lon_idx].values
            lon_slice = slice(lon_value - 1.124/2, lon_value + 1.125/2)
            if mask[lat_idx,lon_idx]:
                total_new[lat_idx, lon_idx] = (population.total.sel(latitude = lat_slice, longitude = lon_slice)).sum()
                count += 1
            else :
                lon_near = lon_array.isel(latitude = slice(lat_idx-1,lat_idx+2), longitude = slice(lon_idx-1,lon_idx+2)).values
                lat_near = lat_array.isel(latitude = slice(lat_idx-1,lat_idx+2), longitude = slice(lon_idx-1,lon_idx+2)).values                
                new_matrix = np.stack((lat_near,lon_near),axis=2)
                new_matrix = new_matrix.reshape(int(np.size(new_matrix)/2), 2)
#                 new_matrix = [new_matrix[idx] for idx in range(3)]
                if np.nansum(new_matrix.flatten()) != 0:
#                     print(new_matrix)
                    count1 += 1
                else :
                    count2 += 1
    
    
    print(count)
    print(count1)
    print(count2)
    print(total_new.sum().values)

regridd_with_nearest(data, population)
```
